```python
import numpy as np
from scipy.optimize import minimize

# Compute an indicative average total weekly advertising budget
avg_w_inv_digital = np.array(digital['digital_investments']).mean()
avg_w_inv_tv = np.array(tv['INVESTMENTS TV']).mean()


# Assuming budget_w is the total average weekly budget for all channels
budget_w = avg_w_inv_digital + avg_w_inv_tv

# Relax the regularization term
regularization_param = 0.01  # Reduced to encourage more diversity

# Define the objective function
def objective_function(allocation_percentages):
    total_revenue = 0
    regularization_penalty = 0
    for tactic, percentage in zip(approximators_xgb.keys(), allocation_percentages):
        tactic_investment = budget_w * percentage
        estimated_sales = approximators_xgb[tactic].predict(np.array(tactic_investment).reshape(-1, 1))
        total_revenue += estimated_sales[0]
        regularization_penalty += regularization_param * (percentage - (1/len(approximators_xgb))) ** 2
    return -(total_revenue - regularization_penalty)

# Modify the initial guess
initial_guess = np.random.dirichlet(np.ones(len(approximators_xgb)), size=1)[0]

# Budget constraint
def budget_w_constraint(allocation_percentages):
    return np.sum(allocation_percentages) - 1

num_channels = len(approximators_xgb)
bounds = [(0.05, 0.3)] * num_channels

constraints = [{'type': 'eq', 'fun': budget_w_constraint}]
result = minimize(objective_function, initial_guess, bounds=bounds, constraints=constraints)
```
